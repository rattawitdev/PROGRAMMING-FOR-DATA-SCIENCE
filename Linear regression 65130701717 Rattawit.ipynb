## Training a Linear Regression Model

### define training data set and test data set

```python
from sklearn.model_selection import train_test_split
X = boston_df[['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','RAD','TAX','PTRATIO','B','LSTAT']]
y = boston_df['DIS']
#=> test_size=0.4 คือ Test 40% Train 60%
#random_state=101 การ random จะได้ผลการทดสอบเหมือนกัน จากการสุ่มแบบเดียวกัน
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

### create model

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

### View the result

```python
print(lm.intercept_)
```

```python
print(lm.coef_)
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

### Predict

```python
predictions = lm.predict(X_test)

import matplotlib.pyplot as plt
plt.scatter(y_test,predictions)
```

```python
sns.distplot((y_test-predictions),kde=False,bins=10)
```

### Calculate error

```python
import numpy as np
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
