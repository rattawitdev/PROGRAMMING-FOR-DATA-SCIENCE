### Question 1 : เดือนที่ดีที่สุดสำหรับการขายคือ และเดือนนั้นได้เงินเท่าไหร่

```python
# ดูว่ามี Record ไหนเป็น Null มั้ย
nan_df = df[df.isna().any(axis = 1)]
nan_df.head()
```

```python
df['Month'] = df['Date'].str[0:1]
df['Month'] = df['Month'].astype('int32')
df.info()
```

```python
df['Sales'] = df['Unit price'] * df['Quantity']
df
```

```python
df.groupby(['Month']).sum()
```

```python
import matplotlib.pyplot as plt

months = range(1, 4)

results = df.groupby(['Month']).sum()['Sales']

plt.bar(months, results)

plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month')

plt.show()
```

### Question 2 : เมืองอะไรมียอดขายสูงที่สุด

```python
df.groupby(['City']).sum()

city_sales_df = df.groupby(['City']).sum()['Sales']

city_sales_df.to_frame().plot.bar()
```

### Question 3 : เวลาใดที่น่าจะเป็นเวลาที่ดีที่สุดในการแสดงโฆษณาให้ลูกค้าซื้อของเพิ่มขึ้น

```python
def convertToDateTime(date,time):
    return pd.to_datetime(date + ' ' + time)

df['Order Date Time'] = df.apply(lambda row: convertToDateTime(row["Date"],row["Time"]), axis=1)

df['Hour'] = df['Order Date Time'].dt.hour
df['Minute'] = df['Order Date Time'].dt.minute
df
```

```python
time_sales_df = df.groupby(['Hour']).sum()['Sales'].to_frame()
time_sales_df 

plot = time_sales_df.plot.line()
```

### Question 4 : กลุ่มผลิตภัณฑ์ใดที่ถูกขายบ่อยที่สุด

```python
product_pair_df = df.groupby(['Product line']).sum()['Quantity']
product_pair_df
product_pair_df.plot.bar(x = 'Product line')
```

### Question 5 : ดูการซื้อของในแต่ละกลุ่มสินค้า ว่ามีผู้ชายหรือผู้หญิง เป็นผู้ซื้อมากกว่ากัน และแสดงผลเป็น Graph

```python
product_female = df[df["Gender"] == 'Female'].groupby(['Product line']).sum()['Quantity'].to_frame()
product_female

product_male = df[df["Gender"] == 'Male'].groupby(['Product line']).sum()['Quantity'].to_frame()
product_male


import matplotlib.pyplot as plt

fig = plt.figure()

ax = product_female.rename(columns={"Quantity": "Female"}).plot.line(figsize=(12,6), style='g-')
product_male.rename(columns={"Quantity": "Male"}).plot.line(figsize=(12,6), style='r-', ax=ax)

plt.show()
```

### การแสดงภาพความสัมพันธ์ในชุดข้อมูล ด้วย heatmap เพื่อไปใช้ในการทำนาย Linear regression ต่อไป

```python
import matplotlib.pyplot as plt
import seaborn as sns

corr_matrix = df.corr()

plt.figure(figsize=(10, 10), facecolor='w', edgecolor='k')
sns.set(font_scale=1.2)
sns.heatmap(corr_matrix,cmap='coolwarm',
            center = 0, 
            annot=True,
            fmt='.1g'
            )
            
```
